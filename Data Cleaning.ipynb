```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
import numpy as np
result=result.replace(np.nan,' ')
```

```python
print(result.shape)
```

```python
result.isnull().sum()
```

```python
corpus_title=[]
for j in result['TITLE']:
    title = re.sub('[^a-zA-Z]', ' ', j)
    title=title.lower()
    title=title.split()
    ps=PorterStemmer()
    all_stopwords=stopwords.words('english')
    all_stopwords.remove('not')
    title=[ps.stem(word) for word in title if not word in set(all_stopwords)]
    title= ' '.join(title)
    corpus_title.append(title)
```

```python
corpus_description=[]
for j in result['DESCRIPTION']:
    title = re.sub('[^a-zA-Z]', ' ', j)
    title=title.lower()
    title=title.split()
    ps=PorterStemmer()
    all_stopwords=stopwords.words('english')
    all_stopwords.remove('not')
    title=[ps.stem(word) for word in title if not word in set(all_stopwords)]
    title= ' '.join(title)
    corpus_description.append(title)
```

```python
corpus_bullet=[]
for j in result['BULLET_POINTS']:
    title = re.sub('[^a-zA-Z]', ' ', j)
    title=title.lower()
    title=title.split()
    ps=PorterStemmer()
    all_stopwords=stopwords.words('english')
    all_stopwords.remove('not')
    title=[ps.stem(word) for word in title if not word in set(all_stopwords)]
    title= ' '.join(title)
    corpus_bullet.append(title)
```
